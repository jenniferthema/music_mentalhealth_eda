# VII) Summary

After investigating the dataset, I can conclude that the age range of respondents participating in the suvey is mostly dominated by those aged between 10 to 26. The respondents' favorite music genres include rock with 24.2% of the total votes and also pop with 15.7% of the total votes, wheareas gospel (0.6%) and latin (0.5%) are two genres with the least amount of votes. The majority of people uses Spotify as their primary music streaming services with 64.3% users, followed by Youtube Music with 11.9% users. A glimpse of background information related to the respondent's mental health can be seen through the graphs provided. Anxiety is the most common and serious mental health issue the respondents may have with around 103 people rating their anxiety scale between 9 and 10. On the other hand, OCD is the least common mental health issue that the respondents have with around 206 people rating their OCD scale between 0 and 1. Based on the results, the effectiveness of music towards mental health is quite apparent with 75.5% of the respondents stating it has improved their mental heatlth.
